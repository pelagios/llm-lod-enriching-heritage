# Second, load the files you want to work with then run the converter

You can upload the files you want to work with to Colab (folder on the left-hand side), then update the filename and folder between the "" bits below. Or you can put in a direct URL if your files are online already.

```python
source = "https://upload.wikimedia.org/wikipedia/commons/2/23/The_excavations_at_Dura-Europos_Final_report._4%2C4%2C1%2C_Pierced_bronzes%2C_enameled_bronzes%2C_and_fibulae.pdf"  # document per local path or URL
converter = DocumentConverter()
```

```python
result = converter.convert(source)
```

# Third, review the results

Once the process above has finished running, run the code below to review the output. How accurately has it transcribed text? Has it correctly structured information in tables or using other formats? Has it recognised handwritten or pencilled text? You may have other questions, depending on your source documents.

```python
print(result.document.export_to_markdown())
```

# Finally, save out the results

Run the code below to export (save) the results of the processes above. If you don't want the filename to be named 'output.md' then edit the text between the double quotes.

You can export into other formats than markdown by changing the format in result.document and changing the file extension e.g. export_to_markdown and output.txt

```python
with open("output.txt", "w") as f:
    f.write(result.document.export_to_text())
```
